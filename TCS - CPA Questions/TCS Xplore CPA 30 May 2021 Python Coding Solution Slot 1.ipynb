```python
class Citizen :
    def __init__(self, citizenID, citizenName, citizenAge,isFrontineWorker,vaccineOptedFor):
        self.citizenID = citizenID
        self.citizenName = citizenName
        self.citizenAge = citizenAge
        self.isFrontineWorker = isFrontineWorker
        self.vaccineOptedFor = vaccineOptedFor
        
        if self.isFrontineWorker=="yes" or self.citizenAge>45:
            self.preference = int(1)
        else:
            self.preference = int(0)
            
class VaccinationDrive:
    def __init__(self,citizen_list):
        self.citizen_list = citizen_list
        
    def getPreferredVaccinationCOunt(self):
        count=0
        for i in self.citizen_list:
            if i.preference == 1:
                count+=1
                
        return count
    
    def getCitizensAsPerVaccine(self, s):
        ordered_list = []
        for i in self.citizen_list:
            if i.vaccineOptedFor == s:
                ordered_list.append(i)
                
        if len(ordered_list) == 0:
            return None
         
        return sorted(ordered_list, key = lambda x:x.citizenAge)
    
  
    
n = eval(input())
citizens = [] 

for i in range(n):
    cid = eval(input())
    name = input()
    age =eval(input())
    isFrontline = input()
    vaccine = input()
    citizen = Citizen(cid, name, age, isFrontline, vaccine)
    citizens.append(citizen)
    
    
vaccine_name = input()

obj2 = VaccinationDrive(citizens)
count = obj2.getPreferredVaccinationCOunt()
citizen_list = obj2.getCitizensAsPerVaccine(vaccine_name)
    

if citizen_list == None:
    print("Citizen not found")
else:
    for i in citizen_list:
        print(i.citizenID)
        print(i.citizenName)
        print(i.citizenAge)
        
if count == None:
    print("Preferred Citizen not found")
else:
    print(count)
    
```
