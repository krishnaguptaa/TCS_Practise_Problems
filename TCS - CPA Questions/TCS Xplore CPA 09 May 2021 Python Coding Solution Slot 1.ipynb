```python
class Painting:
    def __init__(self, pID, pName, pPrice, pType):
        self.pID = pID
        self.pName = pName
        self.pPrice = pPrice
        self.pType = pType
        
class ShowRoom:
    def __init__(self, pList):
        self.pList = pList
        
    def getTotalPaintingPrice(self, ptype):
        price = 0
        for i in self.pList:
            if i.pType == ptype:
                price += i.pPrice
                
        return price
    
    
    def getPainterWithMaxCountOfPaintings(self):
        painter_dict = {}
        
        for i in self.pList:
            if i.pName not in painter_dict:
                painter_dict[i.pName] = 1
                
            else:
                painter_dict[i.pName] = painter_dict[i.pName]+1
                
        name_list = list(painter_dict.keys())
        count_list = list(painter_dict.values())
        
        name_list.sort()
        count_list.sort(reverse=True)
        
        max_count = count_list[0]
        
        for i in name_list:
            if painter_dict[i] == max_count:
                return i
            
            
n = eval(input())
painters = []

for i in range(n):
    pID = eval(input())
    pName = input()
    price = eval(input())
    pType = input()
    
    painter = Painting(pID, pName, price, pType)
    painters.append(painter)

    
painting_type = input()
obj = ShowRoom(painters)

# price of given paintings
price = obj.getTotalPaintingPrice(painting_type)
if price==0:
    print("Painting not found")
else:
    print(price)
    

# painter with max paintings sorted in alphabetical order
painter = obj.getPainterWithMaxCountOfPaintings()
print(painter)      
```
