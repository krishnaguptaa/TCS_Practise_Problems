```python
class Boutique:
    def __init__(self, bId, bName, bType, bRating, points):
        self.bId = bId
        self.bName = bName
        self.bType = bType
        self.bRating = bRating
        self.points = points
        
    
    
class OnlineBoutique:
    def __init__(self, bDict):
        self.bDict = bDict
        
    def getboutique(self, lower, upper, extraPoint, btype):
        
            boutiques = self.bDict[btype]
            valid_boutiques = []
            for boutique in boutiques:
                if boutique.bRating>=lower and boutique.bRating<=upper:
                    boutique.points += extraPoint
                    valid_boutiques.append(boutique)
                    
            valid_boutiques.sort(key=lambda x:x.points, reverse=True)
            return valid_boutiques
    
n = int(input())
boutiques = []
bDict = {}

for i in range(n):
    bID = eval(input())
    bName = input()
    bType = input()
    bRating = eval(input())
    points = eval(input())
    
    boutique = Boutique(bID, bName, bType, bRating, points)
    boutiques.append(boutique)
    
for i in boutiques:
    bDict[i.bType] = []
    
for i in boutiques:
    bDict[i.bType].append(i)
    
    
lower = eval(input())
upper = eval(input())
extrapoint = eval(input())
btype = input()
    
    
    
obj = OnlineBoutique(bDict)

ans = obj.getboutique(lower, upper, extrapoint, btype)

if len(ans)==0:
    print("No boutique found")
else:
    for i in ans:
        print(i.bId, i.bName, i.points )
```
