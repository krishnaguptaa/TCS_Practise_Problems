```python
class Table:
    def __init__(self, tNo, wName, status):
        self.tNo = tNo
        self.wName = wName
        self.status = status
        
    
    
def findWaiterWiseTotalNoOfTable(tList):
    dic = {}
    for i in tList:
        if i.wName not in dic:
            dic[i.wName] = 1
        else:
            dic[i.wName] = dic[i.wName]+1
    
    return dic

def findWaiterNameByTableNo(tList, no):
    for i in tList:
        if i.tNo == no:
            return i.wName


n = int(input())
tables = []
for i in range(n):
    tNo = int(input())
    wName = input()
    status = input()
    tables.append(Table(tNo, wName, status))

tableno = int(input())

# waiters with total tables 
dic = findWaiterWiseTotalNoOfTable(tables)
for i in sorted(dic.keys()):
    print(i,"-",str(dic[i]))

# waiter with table no
waiter = findWaiterNameByTableNo(tables, tableno)
if (waiter==None):
    print("No waiter found")
else:
    print(waiter)
```
