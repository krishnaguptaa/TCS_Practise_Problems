```python
class Sponser:
    def __init__(self, sponserName, sponserCompany,subsidiaries ,sponserCategory):
        self.sponserName = sponserName
        self.sponserCompany = sponserCompany 
        self.subsidiaries = subsidiaries 
        self.sponserCategory = sponserCategory
        
        

        
def getSponser(sponser_list, category):
    new_list = []
    for i in sponser_list:
        if i.sponserCategory == category:
            new_list.append(i)
            
    if len(new_list)==0:
        return None
    return sorted(new_list, key=lambda x:x.sponserName)
    
    
def SponserWithMaximumSubsidiaries(sponser_list):
    sponser_list.sort(key = lambda x:len(x.subsidiaries), reverse=True)
    
    ans = sponser_list[0].sponserName
    
    return ans

    
n = eval(input())
Sponsers = []

for i in range(n):
    sponserName = input()
    sponserCompany = input()
    subsidiaries = []
    m = int(input())
    for j in range(m):
        a = input()
        subsidiaries.append(a)
        
    sponserCategory = input()
    
    Sponsers.append(Sponser(sponserName, sponserCompany, subsidiaries, sponserCategory))

   
    
sponserCategories = input()


# sorted_List of sponsers according name
sorted_list = getSponser(Sponsers, sponserCategories)

if sorted_list == None:
    print("No matching record found")
else:
    for i in sorted_list:
        print(i.sponserName)
    
    
# Sponser name of sponser having maximum subsidiaries
max_subsidiaries = SponserWithMaximumSubsidiaries(Sponsers)
print(max_subsidiaries)
```
