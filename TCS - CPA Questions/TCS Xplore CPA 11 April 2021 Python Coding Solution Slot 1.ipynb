```python
class Doctor:
    def __init__(self, dId, dName, deptName, fee, sunday):
        self.dId = dId
        self.dName = dName 
        self.deptName = deptName
        self.fee = fee
        self.sunday = sunday
        
class Hospital:
    def __init__(self, doctors):
        self.doctors = doctors
        
    def getDoctorList(self, dept):
        doctor_list = []
        for i in self.doctors:
            if i.deptName.lower() == dept.lower() and i.sunday.lower()=="available":
                doctor_list.append(i)
                
        return doctor_list
    
    
    def selectDoctor(self, fee):
        flag = False
        for i in self.doctors:
            if i.fee>fee and i.sunday.lower()=="not available":
                self.doctors.remove(i)
                flag = True
                
        return flag
    
    
doctors = []   
for i in range(5):
    id = int(input())
    name = input()
    dept = input()
    fee = int(input())
    sunday = input()
    
    doctors.append(Doctor(id, name, dept, fee, sunday))
    
    
deptName = input()
fees  = int(input())


obj = Hospital(doctors)

# available on sunday with given dept
available_doctors = obj.getDoctorList(deptName)

if len(available_doctors)!=0:
    for i in available_doctors:
        print(i.dId)
        print(i.dName)
                
else:
    print("Doctor Not Found")
    
    
    
    
# updated list of doctors according to fee and avaibility
updated_list = obj.selectDoctor(fees)
if updated_list==False:
    print("Returning the original list")
    for i in obj.doctors:
        print(i.dId)
        print(i.dName)
        print(i.fee)
else:
    obj.doctors.sort(key= lambda x:x.fee )
    for i in obj.doctors:
        print(i.dId)
        print(i.dName)
        print(i.fee)
        
        
        
        
```
