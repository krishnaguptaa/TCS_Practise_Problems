```python
class Allium:
    def __init__(self, marketID, alliumType, producingState, price):
        self.marketID=marketID
        self.alliumType=alliumType
        self.producingState=producingState
        self.price=price
        
    
class MyClass:
    def findTypeByState(self,allium_list, s):
        for i in allium_list:
            if i.producingState == s:
                return i.alliumType
        
        return None
    
    def sortByPrice(self,allium_list, price):
        for i in range(len(allium_list)):
            for j in range(len(allium_list)):
                if (allium_list[i]).price<(allium_list[j]).price:
                    temp = allium_list[i]
                    allium_list[i] = allium_list[j]
                    allium_list[j] = temp
        allium_list_2 = []
        for i in range(len(allium_list)):
            if allium_list[i].price< price_:
                allium_list_2.append(allium_list[i])
        
                    
        if len(allium_list_2)==0:
            return None 
        return allium_list_2
            
        
n = int(input())
alliums = []

for i in range(n):
    marketID = int(input())
    alliumType = input()
    producingState = input()
    price = int(input())
    
    allium = Allium(marketID,alliumType,producingState,price)
    alliums.append(allium)
    
obj = MyClass()

state = input()

allium_type = obj.findTypeByState(alliums, state)

if allium_type==None:
    print("State not found")
else:
    print(allium_type)
    
    
price_ = int(input())    
allium_list_sorted = obj.sortByPrice(alliums, price_)

if allium_list_sorted == None:
    print(" No allium found ")
    
else:
    for i in allium_list_sorted:
        print(i.alliumType," ",i.price)


```
