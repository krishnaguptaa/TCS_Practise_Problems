```python
class Person:
    def __init__(self, name, height, weight):
        self.name = name 
        self.height = height
        self.weight = weight
        self.bmi = 0
        self.bmi_status = ""
        
    def calculateBmi(self):
        self.bmi = round(self.weight/(self.height*self.height))
        
        
class Society:
    def __init__(self, bmi_status, person_list):
        self.bmi_status = bmi_status
        self.person_list = person_list
        
        
    def calculateBmiAndStatusByName(self, s):
        flag = False
        
        for i in self.person_list:
            if (i.name.lower()==s.lower()):
                flag = True
                i.calculateBmi()
                print(i.bmi, end=" ")
                
                for k in self.bmi_status.keys():
                    v = self.bmi_status[k]
                    if i.bmi<= v[1] and i.bmi>=v[0]:
                        i.bmi_status = k
                        print(k)
            
        return flag
            
                        
                        
    def removeInvalidPersons(self, num):
        invalid_list = []
        for i in self.person_list:
            i.calculateBmi()
            if i.bmi<num:
                invalid_list.append(i)
        
        return invalid_list
    
    
    
    
n = int(input())
person_list = []
for i in range(n):
    name = input()
    height = int(input())
    weight = int(input())
    
    person_list.append(Person(name,height,weight))
    
bmi_status = {}
n1 = int(input())
for i in range(n1):
    key = input()
    l = int(input())
    u = int(input())
    if(l>u):
        l,u = u,l
    bmi_status[key] = (l,u)



obj = Society(bmi_status, person_list)
name = input()
bmi = int(input())


# status updated                   
flag = obj.calculateBmiAndStatusByName(name)
if flag==False:
    print("No person Exists")



# Invalid persons
status_list = obj.removeInvalidPersons(bmi)
for i in status_list:
    print(i.name, end=" ")
    print(i.bmi)
```
