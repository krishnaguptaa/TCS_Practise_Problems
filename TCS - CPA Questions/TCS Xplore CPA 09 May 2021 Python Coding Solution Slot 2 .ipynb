```python
class Book:
    def __init__(self, bID, bTitle, aName, bType, bPrice, status):
        self.bID = bID
        self.bTitle = bTitle
        self.aName = aName
        self.bType = bType
        self.bPrice = bPrice
        self.status = status
        
class Library:
    def __init__(self, bList):
        self.bList = bList
        
        
    def issueBook(self,bTitle, bType):
        li = self.bList
        for i in self.bList:
            if (i.bTitle == bTitle) and  (i.bType == bType) and (i.status =="available"):
                i.status = "unavailable"
                return True, li
            
        return False, li
    
    def findCostliestBookName(self, aName):
        authors_books = []
        for i in self.bList:
            if i.aName == aName:
                authors_books.append(i)
                
        max_cost = 0
        book_name = ""
        for i in authors_books:
            if i.bPrice>max_cost:
                book_name = i.bTitle
                max_cost = i.bPrice
                
        
        for i in authors_books:
            if i.bTitle == book_name:
                return i
            
        return None

    
    

n = int(input())
books = []

for i in range(n):
    bID = input()
    bTitle = input()
    aName = input()
    bType = input()
    bPrice = int(input())
    status = input()
    
    books.append(Book(bID, bTitle, aName, bType, bPrice, status))
    

bTitle = input()
bType = input()
aName = input()


obj = Library(books)

# Books with status updated and issue
ans, blist = obj.issueBook(bTitle, bType)
if ans == False:
    print("Book not found")
else:
    print("Book issued")
    for i in blist:
        print(i.bTitle,"->",i.status)
    
    
# Costliest Book
book = obj.findCostliestBookName(aName)
if book == None:
    print("Author not found")
else:
    print(book.bTitle,"->", book.status)
```
