```python
class Document:
    def __init__(self, docid, docname, docdetails):
        self.docid = docid
        self.docname = docname
        self.docdetails = docdetails
        
        
class DocumentArchive:
    def __init__(self, archivedId, documentList):
        self.archivedId = archivedId
        self.documentList = documentList
        
    def findDateFromDocumentDetails(self):
        list_dates = []
        for i in self.documentList:
            temp = i.docdetails.split(",")
            for j in temp:
                j = j.strip().lower()
                if j[2]=="/" and j[5]=="/":
                    list_dates.append((i.docid, j))
                    
        return list_dates
                     
    
    def countDocumentsGivenType(self, doctype):
        count = 0
        for i in self.documentList:
            details = (i.docname).split(".")
            if details[1] == doctype:
                count+=1
                
        return count
    

    
    
    
    
n = int(input())
documents = []
for i in range(n):
    id_ = int(input())
    name_ = input()
    details_ = input()
    documents.append(Document(id_, name_, details_))
    
        
obj = DocumentArchive(1, documents)

docType = input()

# list of dates
list_dates = obj.findDateFromDocumentDetails()
if len(list_dates) > 0:
    for i in list_dates:
        print(i[0], i[1])
    

    
# count document    
count_doc = obj.countDocumentsGivenType(docType)
print("Document count=",end="")
print(count_doc)
```
